```python
from tkinter import *
root = Tk()
root.title("Ren's project")
root.minsize(500, 100)
root.geometry("800x500")
root.configure(bg="white")
bill = 0
list1=[]
details={"name":"","Address":"","Messenger_notes":"","Card_Number":"","Validity":"","ID":""}

#פונקציות עזר
def add(lbabel, n,prodact):
    global bill#המשתנה הביל הגלובלי שמחוץ לפונקציה
    global list1
    bill += n
    list1.append(prodact)
    lbabel.configure(text="The bill is: " + str(bill))

def remove(lbabel, n,prodact):
    global bill
    global list1
    if (bill-n)>=0:
        bill-=n
        list1.remove(prodact)
        lbabel.configure(text="The bill is: " + str(bill))

def save(name,Address,Messenger_notes):
    global details
    details["name"]=name
    details["Address"]=Address
    details["Messenger_notes"]=Messenger_notes

    
def click5():#חלון ראשון
    global bill
    global list1
    global details
    window5 = Tk()#עוד חלון של רוות
    window5.title("Ren's project")
    window5.minsize(300, 100)
    window5.geometry("400x200")
    window5.configure(bg="white")
    lbl79 = Label(window5, text="You have chosen to pay cash thanks\n appetite", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lbl79.grid(column=0, row=0)
    
def save1(Card_Number,Validity,ID):
    global details
    details["Card_Number"]=Card_Number
    details["Validity"]=Validity
    details["ID"]=ID
    
def click6():#חלון ראשון
    global bill
    global list1
    global details
    window6 = Tk()#עוד חלון של רוות
    window6.title("Ren's project")
    window6.minsize(300, 100)
    window6.geometry("500x300")
    window6.configure(bg="white")
    
    lbl791 = Label(window6, text="You have chosen to pay Credit Card", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lbl791.grid(column=0, row=0)
    
    lblorder3 = Label(window6, text="Card Number:", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblorder3.grid(column=0, row=2)
    lblorder4 = Label(window6, text="Validity:", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblorder4.grid(column=0, row=3)
    lblorder5 = Label(window6, text="ID:", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblorder5.grid(column=0, row=4)
    
    txt4 = Entry(window6,width=10)#תבנית טקסט
    txt4.grid(column=1, row=2)
    txt5 = Entry(window6,width=10)#תבנית טקסט
    txt5.grid(column=1, row=3)
    txt6 = Entry(window6,width=10)#תבנית טקסט
    txt6.grid(column=1, row=4)
    
    btnsave1 = Button(window6, text="save",command=lambda:  save1(txt4.get(), txt5.get(),txt6.get()))
    btnsave1.grid(column=0, row=5)
    btnorder321 = Button(window6, text="close",command=window6.destroy)
    btnorder321.grid(column=0, row=6)

def click():#חלון ראשון
    window1 = Toplevel(root)#עוד חלון של רוות
    global bill
    window1.title("Ren's project")
    window1.minsize(500, 100)
    window1.geometry("800x500")
    window1.configure(bg="yellow")
    
    lblbill = Label(window1, text="The bill is: " + str(bill), font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblbill.grid(column=1, row=1)
    btn1 = Button(window1, text="close",command=window1.destroy)
    btn1.grid(column=0, row=0)
    
    lblmenu = Label(window1, text="Pasta in rose sauce with chicken breast pieces 20$", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblmenu.grid(column=0, row=3)   
    btnmenu =Button(window1, text="Add", command=lambda:  add(lblbill, 20,"Pasta"))
    btnmenu.grid(column=1, row=3)
    btnmenuF =Button(window1, text="remove", command=lambda:  remove(lblbill, 20,"Pasta"))
    btnmenuF.grid(column=2, row=3)

    
    lblmenu1 = Label(window1, text="Tomato Soup 10$", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblmenu1.grid(column=0, row=4)
    btnmenu1 = Button(window1, text="Add", command=lambda:  add(lblbill, 10,"Soup"))
    btnmenu1.grid(column=1, row=4)
    btnmenuF1 =Button(window1, text="remove", command=lambda:  remove(lblbill, 10,"Soup"))
    btnmenuF1.grid(column=2, row=4)
    
    
    lblmenu2 = Label(window1, text="Strawberry cake 5$", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblmenu2.grid(column=0, row=5)
    btnmenu2 = Button(window1, text="Add", command=lambda: add(lblbill, 5,"cake"))
    btnmenu2.grid(column=1, row=5)
    btnmenuF2 =Button(window1, text="remove", command=lambda:  remove(lblbill, 5,"cake"))
    btnmenuF2.grid(column=2, row=5)

def click1():#חלון ראשון
    global list1
    global bill
    window2 = Toplevel(root)#עוד חלון של רוות
    window2.title("Ren's project")
    window2.minsize(500, 100)
    window2.geometry("800x500")
    window2.configure(bg="blue")
    
    lblorder1 = Label(window2, text="your order"+str(list1), font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblorder1.grid(column=0, row=0)
    btnorder1 = Button(window2, text="close",command=window2.destroy)
    btnorder1.grid(column=0, row=1)
    
    lblbill21 = Label(window2, text="The bill is: " + str(bill), font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblbill21.grid(column=0, row=2)
    

    
def click3():#חלון ראשון
    window3 = Toplevel(root)#עוד חלון של רוות
    global bill
    global list1
    global details
    window3.title("Ren's project")
    window3.minsize(500, 100)
    window3.geometry("800x500")
    window3.configure(bg="white")
    lblorder1 = Label(window3, text="your order"+str(list1), font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblorder1.grid(column=0, row=0)
    
    lblorder2 = Label(window3, text="Fill in the details: ", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblorder2.grid(column=0, row=1)
    
    lblorder3 = Label(window3, text="Full Name:", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblorder3.grid(column=0, row=2)
    lblorder4 = Label(window3, text="Address:", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblorder4.grid(column=0, row=3)
    lblorder5 = Label(window3, text="Messenger_notes:", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblorder5.grid(column=0, row=4)
    
    txt = Entry(window3,width=10)#תבנית טקסט
    txt.grid(column=1, row=2)
    txt1 = Entry(window3,width=10)#תבנית טקסט
    txt1.grid(column=1, row=3)
    txt2 = Entry(window3,width=10)#תבנית טקסט
    txt2.grid(column=1, row=4)
    
    btnsave = Button(window3, text="save",command=lambda:  save(txt.get(), txt1.get(),txt2.get()))
    btnsave.grid(column=0, row=5)
    btnorder32 = Button(window3, text="close",command=window3.destroy)
    btnorder32.grid(column=0, row=6)


def click4():#חלון ראשון
    global bill
    global list1
    global details
    window4 = Tk()#עוד חלון של רוות
    window4.title("Ren's project")
    window4.minsize(500, 100)
    window4.geometry("800x500")
    window4.configure(bg="white")
    
    lblorder41 = Label(window4, text="your order"+str(list1), font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblorder41.grid(column=0, row=0)
    lblbill41 = Label(window4, text="The bill is: " + str(bill), font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblbill41.grid(column=0, row=1)
    lblorder411 = Label(window4, text="your details: ", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblorder411.grid(column=0, row=2)
    
    lblname = Label(window4, text="name:"+details["name"], font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblname.grid(column=0, row=3)
    lblAddress = Label(window4, text="your Address: "+details["Address"], font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblAddress.grid(column=0, row=4)
    lblMessenger_notes = Label(window4, text="Messenger_notes:"+details["Messenger_notes"], font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblMessenger_notes.grid(column=0, row=5)    
    lblpayment = Label(window4, text="payment:", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
    lblpayment.grid(column=0, row=6)


    btnpayment = Button(window4, text="Cash",command=click5)
    btnpayment.grid(column=0, row=7)
    btnpayment1 =Button(window4, text="Credit Card",command=click6)
    btnpayment1.grid(column=1, row=7)
    
#עמוד הבית 
lbl = Label(root, text="Welcome to Foox Restaurant\n", font=("Ariel", 30), fg="black", bg="white", relief=RAISED)
lbl1 = Label(root, text='''Hello dear customer, welcome to FOOX restaurant.\n
In front of you is a window for ordering from the restaurant''', font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
lbl2 = Label(root, text="For the menu click here", font=("Ariel", 16), fg="black", bg="white", relief=RAISED)
btn1 =Button(root, text="Menu", command=click)

lbl.grid(column=0, row=0)
lbl1.grid(column=0, row=2)
lbl2.grid(column=0, row=3)
btn1.grid(column=1, row=3)

lbl3= Label(root, text ="To see the order click here", font = ("Ariel",16), fg = "black", bg = "white",relief=RAISED)
btn2 = Button(root, text="Your order", command=click1)
lbl3.grid(column=0, row=4)
btn2.grid(column=1, row=4)

lbl4= Label(root, text ="to finish your order click here", font = ("Ariel",16), fg = "black", bg = "white",relief=RAISED)
btn3 = Button(root, text="click", command=click3)
lbl4.grid(column=0, row=5)
btn3.grid(column=1, row=5)

lbl5= Label(root, text ="For payment click here", font = ("Ariel",16), fg = "black", bg = "white",relief=RAISED)
btn4 = Button(root, text="Payment details", command=click4)
lbl5.grid(column=0, row=6)
btn4.grid(column=1, row=6)
root.mainloop()
exit()
```
